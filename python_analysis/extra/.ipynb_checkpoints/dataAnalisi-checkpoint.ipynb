```python
HIT=[np.array([MAP.iloc[idata]['x'].to_numpy(),MAP.iloc[idata]['y'].to_numpy()]) for idata in nData]
```

```python
typeLab='reco'
lc=None
match typeLab:
    case 'reco':
        lc=('np_reco','ne_reco') 
    case 'gpu':
        lc=('np_gpu','ne_gpu') 
    case 'track':
        lc=('np_track','ne_track') 
    case _:
        raise Exception('Wrong type of label')

label=np.array([DATA[lc[0]],DATA[lc[1]]]).T
```

```python
nP=label.T[0]
ne=label.T[1]
```

```python
_=plt.title("Number of Particles for "+typeLab.upper()+" classification")
_=plt.hist(nP,density=True,bins=range(10),align='left')
plt.vlines([nP.mean()],0,max(_[0]),color="red")
s=f"Mean={nP.mean():.2f}"
__=plt.text(5,0.2,s)
_=plt.xticks(_[1])
plt.show()
_=plt.title("Number of Electron for "+typeLab.upper()+" classification")
_=plt.hist(ne,density=True,bins=range(5),align='left')
plt.vlines([ne.mean()],0,max(_[0]),color="red")
per=_[0]
s=f"Mean={ne.mean():.2f}"
__=plt.text(2,0.8,s)
s=f"Electron percentage:\n0 e = {per[0]:.2%}\n1 e = {per[1]:.2%}\n2 e = {per[2]:.2%}\n"
__=plt.text(2,0.5,s)
_=plt.xticks(_[1])
plt.show()
```

```python
import numpy.random as rd
X=np.empty((11,0)).tolist()
Y=np.empty((11,0)).tolist()
tr=9800
for i,h in enumerate(HIT):
        X[label[i][0]].append(h[0][h[0]<tr])
        Y[label[i][0]].append(h[1][h[1]<tr])
        
for nC in range(len(X)):
    _=plt.title("Circle = "+str(nC))
    t=rd.randint(len(X[nC]))
    _=plt.hist2d(X[nC][t],Y[nC][t],bins=45,range=[[-300,100],[-200,200]])
    plt.show()
```
